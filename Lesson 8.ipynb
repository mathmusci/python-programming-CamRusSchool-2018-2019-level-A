## Графика: черепашки, оставляющие следы на песке

В мире Python черепашкой называют объект, напоминающий обыкновенную черепаху, которая ползает, например, по песку, и оставляет за собой след.
Только в Python это небольшая черная стрелочка, которая медленно перемещается по экрану. Черепашка хорошо подходит для изучения основ компьютерной графики, и в этой главе мы будем рисовать с ее помощью линии и несложные контуры.

Инструменты, необходимые для рисования с использованием такой черепашки находятся в модуле `turtle`. 

### Модуль `turtle`

Давайте посмотрим, как работает turtle. Запустите оболочку Python и импортируйте модуль `turtle`:

```
>>> import turtle
```

### 

Итак, мы импортировали модуль turtle, и теперь нужно создать холст — чистое пространство для рисования, вроде холста у художников.
Для этого вызовем функцию Pen из модуля turtle, и она автоматически создаст холст и черепашку как объект, которым мы можем управлять (и которому нам даже необходимо дать имя!). Для этого введите в оболочке Python такую команду:

```
Bob = turtle.Pen()
```

На экране должно появиться белое окно (это холст) со стрелочкой посередине.

Обратите внимание, что мы используем название объекта `Pen` вместе с названием модуля (`turtle`), который мы загрузили и в котором этот объект `Pen` находится. Для того, чтобы не использовать каждый раз название модуля, когда мы хотим воспользоваться его содержимым (объектами, переменными, константами), небходимо сделать следующее:

```
from turtle import Pen
...
Bob = Pen()
```

### Перемещение черепашки
Пользуясь различными командами модуля `turtle` мы можем заставить черепашку двигаться и оставлять за собой след на песке (графическом полотне). Например, команда forward служит для перемещения черепашки вперед. Чтобы черепашка `Bob` сдвинулась на 50 пикселей, введите команду 

``
Bob.forward(50)
``

Черепашка Bob передвинулась на 50 пикселей. Пиксель - это одна экранная точка, самый маленький элемент изображения. Все, что вы видите на экране монитора, состоит из пикселей — крошечных квадратных точек. Если посмотреть в увеличении на холст и линию, которую нарисовала черепашка, обнаружится, что и след черепашки, и она сама — просто набор пикселей. Это и есть самая простая компьютерная графика.

Теперь повернем черепашку на 90 градусов влево, введя такую команду:

```
Bob.left(90)
```

Если вы еще не знаете, что такое градусы, вот как можно это себе представить. Вообразите, что вы стоите в центре круга.

Направление, в котором вы смотрите, это 0 градусов.
 * Если вы вытянете левую руку вбок, это будет 90 градусов влево.
 * Если вы вытянете вбок правую руку, это будет 90 градусов вправо.

Вот изображение этих 90-градусных поворотов:

![](figures/lesson_8_angles.png)

Если продолжать двигаться по часовой стрелке от вашей правой руки и дальше, 180 градусов — это прямо за вашей спиной. 270 градусов - там, куда указывает левая рука, а 360 градусов - направление вашего взгляда, то есть точка, откуда мы начали. Получается, что градусы проходят полный круг от 0 до 360. Вот круг, размеченный на градусы слева направо, с шагом в 45 градусов:

![](figures/lesson_8_angles2.png)

После того как черепашка повернется влево, она будет смотреть в новом направлении — как если бы вы развернулись на 90 градусов туда, куда показывает ваша левая рука. Поэтому команда `Bob.left(90)` разворачивает стрелочку острием вверх (так как вначале она указывала вправо)!

Кстати, команда `Bob.left(90)` даст тот же результат, что и команда `Bob.right(270)`, а `Bob.right(90)` соответствует `Bob.left(270)`. Просто вообразите себе круг с градусами, и все станет понятно.

Теперь давайте нарисуем квадрат. Добавьте к уже введенным командам вот такие:

```
>>> Bob.forward(50)
>>> Bob.left(90)
>>> Bob.forward(50)
>>> Bob.left(90)
>>> Bob.forward(50)
>>> Bob.left(90)
```
Черепашка должна изобразить квадрат и развернуться:

![](figures/lesson_8_square.png)

Чтобы очистить холст, введите команду `reset`. При этом с холста исчезнет все нарисованное, а черепашка вернется в начальную позицию.

```
>>> Bob.reset()
```

Либо можно использовать команду `clear`. Холст также очистится, но черепашка останется на прежнем месте.

```
>>> Bob.clear()
```

Также черепашку можно поворачивать вправо командой `right` и перемещать назад командой `backward`. Еще есть команда `up`, убирающая перо с холста (иными словами, она говорит черепашке, что ей не надо оставлять за собой след, то есть рисовать). С помощью команды `down` рисование можно снова включить. Вызывать эти функции нужно так же, как и остальные, которыми мы уже пользовались.

Давайте опробуем некоторые из этих команд, создав еще один рисунок. Пусть на этот раз черепашка изобразит две линии. Введите следующий код:

```
>>> Bob.reset()
>>> Bob.backward(100)
>>> Bob.up()
>>> Bob.right(90)
>>> Bob.forward(20)
>>> Bob.left(90)
>>> Bob.down()
>>> Bob.forward(100)
```

Сперва командой `t.reset()` мы очистили холст и переместили черепашку в начальную позицию. Затем с помощью `t.backward(100)` мы переместили ее на 100 пикселей назад и вызовом` t.up()` отключили рисование.

Далее командой `t.right(90)` мы развернули черепашку на 90 градусов вправо так, чтобы она указывала вниз, и командой `t.forward(20)` передвинули ее на 20 пикселей вперед. При этом на экране ничего не появилось, ведь раньше мы отключили рисование. Затем командой `t.left(90)` мы повернули черепашку на 90 градусов влево, после чего она стала указывать вправо, и командой down снова включили рисование. И наконец, командой `t.forward(100)` мы прочертили линию, параллельную первой.

В результате вышел такой рисунок:

![](figures/lesson_8_two_lines.png)

### Что мы узнали
Мы узнали, как начать пользоваться модулем `turtle`, импортировав его сперва. Мы начертили несколько линий с помощью поворотов влево (left) и вправо (right), а также движений вперед (forward) и назад (backward). Выяснили, что рисование можно отключить командой up и включить командой down. Также мы узнали, что для управления поворотами черепашки используются значения в градусах.
