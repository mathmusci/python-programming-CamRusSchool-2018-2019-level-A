# Урок девятый

## Команды модуля `turtle`

Вы уже знаете, что как только мы импортировали модуль `turtle` и создали черепашку используя класс Pen, Python открывает окно вывода графики и помещает в его центр нашу черепашку, ориентированную "на восток", т.е. влево:

```
import turtle
Ben = turtle.Pen()
```

Как вы помните, мы чертим на экране перемещая черепашку с помощью последовательности команд. Давайте перечислим эти команды и добавим к ним команды, которые нам еще не известны.

## Команды движения

|команда|описание|
|:--|:--|
|`forward(distance)`|передвинуть черепашку на расстояние distance в текущем направлении|
|`backward(distance)`|передвинуть черепашку на расстояние distance в направлении противоположном текущему|
|`right(angle)`|повернуть черепашку на угол `angle` по направлению часовой стрелки|
|`left(angle)`|повернуть черепашку на угол `angle` в направлении противоположном ходу часовой стрелки|
|`goto(x,y)`|переместить черепашку в точку с координатами `(x,y)`|
|'home()'|переместить черепашку в начало координат, ориентировав её "на восток"|
|'speed(speed)'|изменить скорость перемещения черепашки - от нуля до десяти (ноль соответствует самой быстрой скорости)|

Необходимо помнить, что положение черепашки в графическом окне можно всегда определить используя следующую систему координат:

![система координат](figures/lesson_9_coords.png)

## Команды рисования

|команда|описание|
|:--|:--|
|`pendown()`|опустить перо на холст - черепашка будет оставлять след при каждом следующем перемещении|
|`penоп()`|поднять перо с холста - черепашка не будет оставлять следов на холсте до тех пор пока мы не опустим перо на холст|
|`pensize(width)`|установить толщину линии, `width` - положительное число|
|`circle(radius, extent=None,steps=None)`|нарисовать окружность или дугу окружности|
|`dot(size=None,color)`|нарисовать точку размера `size` с закрашиванием цвета `color`|
|`stamp`|оставить профиль черепашки; функция возвращает номер `stamp_id` профиля черепашки как объекта|
|`clearstamp(stamp_id)`|удалить объект профиля черепашки со штампом `stamp_id`|
|clear()|очистить экран без перемещения черепашки|
|reset()|очистить экран и переместить черепашку в начало координат (0,0)|

Отдельного внимания заслуживает функция `circle`. С ее помощью можно рисовать дуги окружностей. Для этого используется аргумент `extent`, которому необходимо передать значения центрального угла дуги. Окружности и дуги всегда рисуются слева от черепашки, т.е. местоположение окружности (или дуги) напрямую зависит не только от положения, но и от ориентации черепашки:

```
ben = turtle.Pen()
ben.circle(50, 90)
```
Аргумент `step` функции `circle` указывает на то сколько углов вписанного многогранника будет нарисовано (на самом деле окружность рисуется не идеальная - она представлена как аппроксимация многоугольником):

```
import turtle
ben = turtle.Pen()
ben.left(90)
ben.speed(0)
ben.circle(150, 180, steps=4)
ben.circle(150,180, steps=8)
turtle.done()
```

Теперь мы готовы нарисовать рожицу - создайте новый файл и напечатайте следующий текст программы, а затем исполните ее:

```
import turtle

Ben = turtle.Pen()
Ben.speed(10)

Ben.penup()
Ben.forward(75)

# draw an eye
Ben.pendown()
Ben.right(90)
Ben.circle(25)
Ben.circle(10)

# left side of face
Ben.penup()
Ben.right(90)
Ben.forward(150)

# draw an eye
Ben.pendown()
Ben.right(90)
Ben.circle(25)
Ben.circle(10)

# centre and down
Ben.penup()
Ben.right(90)
Ben.forward(75)
Ben.right(90)
Ben.forward(50)

# draw a nose
Ben.pendown()
Ben.left(45)
Ben.forward(40)
Ben.right(135)
Ben.forward(56.56)
Ben.right(135)
Ben.forward(40)

# centre, down, then 100 left
Ben.penup()
Ben.right(135)
Ben.forward(50)
Ben.right(90)
Ben.forward(100)
Ben.left(90) # need to face east

Ben.pendown()
Ben.circle(100, 180) # smile

turtle.done()
```

## Использование цвета

При рисовании с использованием модуля `turtle` можно менять цвет пера или холста. Можно также рисовать закрашенные фигуры. При этом `turtle` распознает названия основных цветов, например "red", "blue" или "green". Список названий цветов можно найти по следующему адресу: https://www.tcl.tk/man/tcl8.4/TkCmd/colors.htm

## Команды смены цвета

|команда|описание|
|:--|:--|
|`pencolor(color)`|`color` - цвет, строка или RGB кортеж|
|`fillcolor(color)`|`color` - цвет, строка или RGB кортеж|
|`colormode(255)`|установить RGB цветовые коды в диапазоне 0-255|
|`color(pen_color, fill_color)`|установить цвета пера и заливки области|
|`begin_fill()`|указание начать закрашивать область|
|`end_fill()`|указание закончить закрашивать область|

Создайте новый файл и проверьте результат работы следующей программы:

```
from turtle import *

colormode(255)
fillcolor('blue')
begin_fill()
for n in range(4):
    forward(80)
    right(90)
end_fill()
done()
```

## Еще немного полезных команд

|команда|описание|
|:--|:--|
|`shape(a_shape)`|установить профиль (форму) чеперашки на `a_shape` (можно выбрать один из следующих вариантов: "classic", "arrow", "turtle", "triangle", "square", "circle")|
|`write(s, font=font_tuple)`|напечатать строку `s` используя шрифт, детали которого переданы в `font_tuple=(font,size,type)`|
|`onclick(click_fn)`|это так называемый callback - при нажатии мышкой на профиле черепашки запускается функция `click_fn`|
|`onrelease(release_fn)`|функция `release_fn` выполнится при отпускании кнопки мыши|
|`hideturtle()`|спрятать профиль черепашки|
|`showturtle()`|показать профиль черепашки|
