## Цикл while

Цикл `for` - не единственный вид циклов в языке Python. Есть также цикл `while`, который используется, если нужное количество повторов заранее неизвестно, тогда как `for` расчитан на определенное число повторов. Представьте себе лестницу с 20 ступеньками. Это внутренняя лестница в помещении, и вы знаете, что запросто преодолеете ступеньки. Таков и цикл `for`.

```
>>> for step in range(1, 21):
        print(step)
```

А теперь представьте, что вам нужно повторить определенный фрагмент кода несколько (возможно много!) раз, однако вы не знаете заранее сколько раз это нужно сделать. Цикл `while` служит как раз для этих целей. Синтаксис этой конструкции выглядит следующим образом:

```
while condition:
    code
```

Давайте теперь рассмотрим пример и воспользуемся циклом `while`. Вспомните наш код с введением возраста:

```
age = int(input('Input your age: '))
if 7 <= age <= 20:
    print("You are quite young!")
else:
    print("You are old!")
```

Давайте подумаем, что произойдет, если пользователь введет 3 (или 350). Это может произойти по ошибке, и если такая ошибка произошла, мы захотим дать пользователю возможность ввести свой возраст еще раз (и вообще, столько раз сколько это будет небходимо). Вот как это можно сделать с помощью `while`:

```
age = 0
while age < 7 or age > 105:
    age = int(input('Input your age: '))
if 7 <= age <= 20:
    print("You are quite young!")
else:
    print("You are old!")
```

Вот как этот код контролирует ввод:
```
Input your age: 1
Input your age: -5
Input your age: 109     
Input your age: 23
You are old!
```

Практически, `while` циклически запрашивает возраст и не позволяет выполнение кода следующего за `while` пока не будет введен возраст из ожидаемого диапазона.

## Игра в отгадывание числа
Давайте теперь напишем нашу первую настоящую игру! В этой игре компьютер попросит нас загадать число от одного до ста и затем последовательно попытается угадать загаданное нами число - на каждую такую попытку мы сможем только ответить "больше" ("greater") наше число, либо "меньше" ("less"), чем число предложенное компьютером.

```
print("Guess a number between 1 and 100 and I am going to guess it")

initial_guess = 50

m = 1
M = 100

guess = initial_guess
while True:
	answer = ''
	while answer not in {'less', 'greater', 'equal'}:
	    answer = input("Is your number less or greater than %s? \
Please, type less, greater (or equal if my guess is correct): " % guess)

	if answer == 'equal':
		break
	elif answer == 'less':
		M = guess		
	elif answer == 'greater':
		m = guess

	guess = int(round((m+M)/2))

print("Hurray: I have guessed - your number is %s" % guess)
```

Обратите внимание на то, что мы используем служебное слово `break` для того, чтобы выйти из цикла `while`, который в противном случае никогда бы не смог закончиться.
