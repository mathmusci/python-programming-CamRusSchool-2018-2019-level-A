## Строение функции

Функция состоит из трех частей: имени, аргументов и тела. Вот пример простой функции:

```
def testfunc(name):
 ̺ ̺ ̺ ̺print('Hello ' + name + '!')
```

Имя этой функции - `testfunc`. У нее есть единственный аргумент - `myname`, а ее тело - это блок кода, идущий сразу после строки, которая начинается с def (сокращение от `define` - определить). Аргумент - это специальная переменная, которая существует, только пока функция выполняется. Функцию `testfunc` можно запустить (или, как говорят программисты, вызвать), указав ее имя, а после него, в скобках, значение аргумента:

```
>>>testfunc('John')
Hello John!
```

Зачастую из функции полезно вернуть какое-то значение. Это делается с помощью команды `return`. Например, представьте такую функцию для расчета сбережений:

```
def savings(pocket_money, spending):
 ̺ ̺ ̺ ̺return pocket_money – spending
```

Эта функция принимает два аргумента. Вычисляя результат, она вычитает величину `spending` от величины  `pocket_money`. Команда `return` возвращает результат вычислений в ту часть кода, откуда функция была вызвана, и его можно сохранить в переменной (обычным способом, с помощью знака "=") либо вывести на экран:

```
>>> print(savings(10, 5))
5
```

Теперь давайте вернемся к рисованию и создадим функцию, которая рисует квадрат определенного размера. Наша программа может выглядет следующим образом (наберите текст этой программы с отдельном файле):

```
import turtle

def square(turtle_name, side_length):
	for k in range(4):
	    turtle_name.forward(side_length)
	    turtle_name.left(90)

Ben = turtle.Pen()
Ben.left(10)
square(Ben, 100)
turtle.done()
```

Обратите внимание на следующее:

1. Функция `square` принимает два аргумента: turtle_name (имя нашей черепашки, которая и будет рисовать квадрат) и side_length (размер квадрата).
2. Мы рисуем квадрат повторением последовательности действий "прямо на `side_length` шагов" и затем "налево" четыре раза подряд.

А что если мы хотим нарисовать квадрат левый нижний угол которого находится в точке с координатами (x,y)? Посмотрите как можно изменить текст программы которую мы уже создали, чтобы достичь этого: 

```
import turtle

def square(turtle_name, x, y, side_length):
	turtle_name.penup()
	turtle_name.goto(x,y)
	turtle_name.pendown()
	for k in range(4):
	    turtle_name.forward(side_length)
	    turtle_name.left(90)
	
Ben = turtle.Pen()
Ben.speed(0)

square(Ben, 100, 100, 50)
square(Ben, -100, 100, 50)
square(Ben, -150, -150, 350)
square(Ben, -100, -100, 50)
square(Ben, 100, -100, 50)

turtle.done()
```
